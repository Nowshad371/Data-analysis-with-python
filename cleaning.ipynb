```python
'''There are several approches for normalization
    
   1. simple Feature Scaling
   
   x(new) = x(old)/x(max) {this make new values range between 0 - 1}
   
   2. Min-Max
   
   x(new) = x(old)-x(min) / x(max)-x(min) {this make new values range between 0 - 1}
   
   3. Z-score /standard score
   
   x(new) = x(old) - mean/standard {the resulting value around 0 and tipically range between negative 3 and positive 3 
                                   and can be higher or lower}
   

'''
```

```python
#normalization

#1. simple Feature Scaling

df['education'] = df['education']/df['education'].max()
```

```python
df.head()
```

```python
# min-max method
 
df['education'] = (df['education'] - df['education'].min())/(df['education'].max() - df['education'].min())
df.head()
```

```python
#Z-score method

df['education'] = (df['education'] - df['education'].mean())/df['education'].std()

df.head()
```

```python
'''Binning - 1. It is grouping of values into bins. for example we can bin age from (0-5,6-10,11-15 etc.) 
                sometimes it can improve accuracy of the predictive models.Sometimes we use data binning to group a set of 
                numerical values into smaller number of bins to have better understanding of the data distribution.
                
                
             2. helps to convert numeric into categorical variables.
             
             
             3. Group a set of numerical values into a set of bins


'''
```

```python
bins = np.linspace(min(df['yearsmarried']),max(df['yearsmarried']),5)

groups_names = ['Low','Low-Medium','Medium','High']

df['yearsmarried_Binned'] = pd.cut(df['yearsmarried'],bins,labels =groups_names,include_lowest = True)
df.head()
```

```python
#For visualizing binned data we can use histrogram

import matplotlib.pyplot as plt
%matplotlib inline

plt.hist(df['yearsmarried_Binned'], 3,
         density=True,
         histtype='bar',
         facecolor='b',
         alpha=0.5)

plt.show()
```
