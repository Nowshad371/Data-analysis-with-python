###### Group By

With the help of group by method we can build the relation between several columns

In pandas this can be done by using df.groupby() method

1. Group by method can be apply in categorical variables
2. Group data into different subset using categories of that data
3. we can group by single or multiple variables


Example: Lets say we want to find average price of different vechicle and
         observe how the differ between body style and drive wheels variable

###### pearson Correlation

One of the way to measure strength of the correlation between continous numerical variables is by a method known as Pearson correlation

Pearson Correlation gives two values:

    1. Correlation coefficient
        
        * Close to +1: Large Positive relationship
        * Close to -1: Large Negative relationship
        * Close to  0: No Relationship
    
    2. P-value
      
        * P-value <0.001 Strong certainty in the result
        * P-value <0.05  Moderate certainty in the result
        * P-value <0.01  Weak certainty in the result
        * P-value >0.1   No certainty in the result
        


   Strong Correlation:
   
   * Correlation coefficient Close to 1 or -1
   * P-value <0.001 
   
   
 Lets look at the pearson correlation between two variables (horsepower,price)
 
 pearson_coef, p_value = stats.pearsonr(df['horsepower'],df['price'])
 
 here correlation of both variable are:
 
 * Pearson correlation: 0.81 (value is 0.81 which is close to 1 so this is a strong correlation)
 
 * P-value : 9.35 e-48 
  (P-value is much smaller than 0.001 so we can conclude,we are certain about the strong positive correlation]
